```python
import numpy as np
```

```python
# 1.Write a NumPy program to find the sum of all elements in an array [10, 20, 30]

arr = np.array([10, 20, 30])

print("Array:", arr)
print("Sum:", np.sum(arr))
```

```python
# 2.Write a NumPy program to find the shape of the array
#[[1, 2, 3], [4, 5, 6]].

arr2 = np.array([[1, 2, 3], [4, 5, 6]])

print("Array:\n", arr2)
print("Shape:", arr2.shape)
```

```python
# 3.Write a NumPy program to find the maximum element in [5, 12, 3, 9].

arr3 = np.array([5, 12, 3, 9])

print("Array:", arr3)
print("Maximum:", np.max(arr3))
```

```python
# 4.Write a NumPy program to add two arrays
#[1, 2, 3] and [4, 5, 6].

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = a + b

print("Array 1:", a)
print("Array 2:", b)
print("Addition:", result)
```

```python
# 5.Write a NumPy program to check the data type of an array [1, 2, 3].

arr4 = np.array([1, 2, 3])

print("Array:", arr4)
print("Data type:", arr4.dtype)
```
