```python
# 1. Area of circle
def area_circle(r):
    return 3.14 * r * r

r = float(input('Enter radius: '))
print('Area =', area_circle(r))
```

```python
# 2. Largest of three numbers
def largest(a,b,c):
    return max(a,b,c)

a = int(input('Enter first number: '))
b = int(input('Enter second number: '))
c = int(input('Enter third number: '))
print('Largest =', largest(a,b,c))
```

```python
# 3. Length of string
def str_length(s):
    count = 0
    for _ in s:
        count += 1
    return count

s = input('Enter string: ')
print('Length =', str_length(s))
```

```python
# 4. Prime check
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

num = int(input('Enter number: '))
print('Prime' if is_prime(num) else 'Not Prime')
```

```python
# 7. Sum 1 to n
def sum_n(n):
    total = 0
    for i in range(1,n+1):
        total += i
    return total

n = int(input('Enter n: '))
print('Sum =', sum_n(n))
```

```python
# 8. Count even numbers
def count_even(lst):
    count = 0
    for i in lst:
        if i % 2 == 0:
            count += 1
    return count

lst = list(map(int, input('Enter numbers: ').split()))
print('Even count =', count_even(lst))
```

```python
# 9. Primes 1 to 10
print('Prime numbers:')
for num in range(1,11):
    if num > 1:
        for i in range(2,num):
            if num % i == 0:
                break
        else:
            print(num)
```
