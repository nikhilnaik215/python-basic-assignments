```python
# 1. Area of circle
def area_circle(r):
    return 3.14*r*r
print(area_circle(5))
```

```python
# 2. Largest of three numbers
def largest(a,b,c):
    return max(a,b,c)
print(largest(3,7,5))
```

```python
# 3. Length of string without built-in
def str_length(s):
    count=0
    for _ in s:
        count += 1
    return count
print(str_length('hello'))
```

```python
# 4. Prime check
def is_prime(n):
    if n<2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n%i == 0:
            return False
    return True
print(is_prime(7))
```

```python
# 5. Fibonacci (recursive)
def fib(n):
    if n<=1:
        return n
    return fib(n-1) + fib(n-2)
print(fib(6))
```

```python
# 6. Reverse string (recursive)
def reverse(s):
    if len(s) == 0:
        return s
    return reverse(s[1:]) + s[0]
print(reverse('hello'))
```

```python
# 7. Sum 1 to n (for loop)
def sum_n(n):
    total = 0
    for i in range(1,n+1):
        total += i
    return total
print(sum_n(5))
```

```python
# 8. Count even numbers in list
def count_even(lst):
    count = 0
    for i in lst:
        if i%2 == 0:
            count += 1
    return count
print(count_even([1,2,3,4,6]))
```

```python
# 9. Prime numbers between 1 and 10
for num in range(1,11):
    if num>1:
        for i in range(2,num):
            if num % i == 0:
                break
        else:
            print(num)
```
